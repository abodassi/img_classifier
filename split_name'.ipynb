```python
import os
import re
import pandas as pd
import numpy as np
train_data = "C:/Users/abd/Desktop/cars/img/data/train"

brands_names = []
if os.path.exists(train_data):
    for file_name in os.listdir(train_data):
        match = re.findall(r'\b\w+$', file_name)
        brands_names.append(match)
    n = np.array(brands_names).flatten()
    df = pd.Series(n).unique()
    print(df)
            
else:
    print(f"The folde does not exist.")
base_path = "C:/Users/abd/Desktop/car_train"
```

```python
import shutil
def newdir(images_base_path,base_path):
    for brand_folder in os.listdir(images_base_path):
        brand_folder_path = os.path.join(images_base_path, brand_folder)
        
        if os.path.isdir(brand_folder_path):
            brand_name = re.findall(r'\b\w+$', brand_folder).pop() 
            
            brand_folder_path_new = os.path.join(base_path, brand_name)
            
            if not os.path.exists(brand_folder_path_new):
                os.makedirs(brand_folder_path_new)
                print(f"Created folder for brand: {brand_name}")
            
            for file_name in os.listdir(brand_folder_path):
                if file_name.endswith(('.jpg', '.jpeg', '.png', '.gif')):  
                    source_file = os.path.join(brand_folder_path, file_name)
                    destination_file = os.path.join(brand_folder_path_new, file_name)
                    
                    shutil.copy(source_file, destination_file)
                    print(f"Copied {file_name} to {brand_folder_path_new}")
newdir(train_data,base_path)
```
